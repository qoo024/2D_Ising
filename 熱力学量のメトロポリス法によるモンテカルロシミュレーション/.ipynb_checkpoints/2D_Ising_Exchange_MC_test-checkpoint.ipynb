```python
from random import random, randrange
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
#与えられたスピン配位alisに対してエネルギーを計算する関数を定義
##参考文献のコードよく分からない！
def Ecalc2(alis2, Nx, Ny):
    dum = 0
    
    #右側のスピンとの相互作用
    for i in range(0, Nx):
        for j in range(0, Ny):
            
            migi = i + 1
            
            if migi == Nx:
                migi = 0
            dum += alis2[i,j]*alis2[migi,j]
    
    #上側のスピンとの相互作用
    for i in range(0, Nx):
        for j in range(0,Ny):
            
            ue = j+1
            
            if ue == Ny:
                ue = 0
            
            dum += alis2[i,j]*alis2[i,ue]
            
    return dum

#ランダムな初期状態生成 : アップスピンとダウンスピンは同じ。M=0
##ランダムでNN/2のスピンを選んでひっくり返しているが重複している可能性がある！
def Initial_rand(s, Nx, Ny):
    NN = int(Nx*Ny/2)
    for k in range(NN):
        i = randrange(Nx) #0からN-1まででランダムな数が選ばれる。
        j = randrange(Ny)
        s[i,j] = -1*s[i,j]
        
    return s

Nx = 20
Ny = 20
Ntot = Nx * Ny

KBT_lis = np.linspace(0.001, 6, 201) #温度を0.001から6まで201刻みで変動させる
ETplot = []
MTplot = []
CTplot = []

for KBT in KBT_lis:
    J = 1 #交換相互作用
    B = 0.0 #外部磁場
    steps = 40000 #MCステップ
    avsteps = int(steps/5) #最後のsteps/5回分だけを採用し、その前はバーンアウト(?)
    
    #初期状態生成:ランダムスピン配置
    s = np.ones([Nx,Ny], int) #Nスピン分を全てアップスピン(+1)としてセット
    
    s = Initial_rand(s, Nx, Ny)
    
    E = -J * Ecalc2(s, Nx, Ny) - B * np.sum(s) #初期エネルギー
    E2 = E**2 #初期エネルギーの二乗を格納
    
    #セットアップ
    eplot = []   #エネルギー値を格納 
  #  Eavplot = [] #エネルギーの熱平均値を格納
    e2plot = [] #エネルギー値の二乗を格納
    
    eplot.append(E/Ntot)
  #  Eavplot.append(E/Ntot)
    
    e2plot.append((E/Ntot)**2)
    
    Cvplot = [] ##各温度kBTの比熱を格納
    
    
    M = np.sum(s) #磁化
    Mplot =[]
    Mavplot = []
    Mplot.append(M/Ntot)
   # Mavplot.append(M/Ntot)
    
    #メイン
    
    for k in range(steps):
        i = randrange(Nx-1)
        j = randrange(Ny-1)
        
        s_trial = s.copy()   #deep copyに相当
        s_trial[i,j] = -1*s[i,j]
        delta_E = 2*s_trial[i,j]*(-1*J)*(s[i+1,j]+s[i-1,j]+s[i,j+1]+s[i,j-1])-B*(s_trial[i,j]-s[i,j])
        
        #メトロポリス法による状態更新
        if delta_E < 0:
            s = s_trial
            E = E + delta_E
        else:
            if random() < np.exp(-delta_E/KBT):
                s = s_trial
                E = E + delta_E
        
        eplot.append(E/Ntot)
        e2plot.append((E/Ntot)**2)
        M = np.sum(s)
        Mplot.append(M/Ntot)
    
    ETplot.append(np.sum(eplot[-avsteps:])/avsteps)
    MTplot.append(np.sum(Mplot[-avsteps:])/avsteps)
    CTplot.append((np.sum(e2plot[-avsteps:])/avsteps - (np.sum(eplot[-avsteps:])/avsteps)**2)/KBT**2)
    
plt.plot(KBT_lis, ETplot)
plt.ylabel("Energy per spin")
plt.xlabel("$k_{B}T$")
plt.show()
    
plt.plot(KBT_lis,MTplot)
plt.ylabel("magnetic moment per spin")
plt.xlabel("$k_{B}T$")
plt.show()

plt.plot(KBT_lis,CTplot)

plt.ylabel("Heat capacity per spin")
plt.xlabel("$k_{B}T$")
plt.show()
```
